```python
from mpl_toolkits import mplot3d
import random
import matplotlib.pyplot as plt
import numpy as np
from numpy import linalg as la
```

```python
f=lambda x: 2*x+ 1
```

```python
def gen_data(n):
    x = np.linspace(0,1,n)
    y = np.zeros_like(x)
    eps=1E-1
    for i in range(n):
        y[i] = f(x[i]+eps**random.randint(0,10))

    return x, y
```

```python
def Ec_normales(n):
  x,y=gen_data(n)
  A=np.zeros((len(y),2))
  A[:,0]=1.0
  A[:,1]=x
  b=y

  At_A=A.T@A
  At_b=A.T@b
  Params=SolveChol(At_A,At_b)

  z=Params[0]+Params[1]*x

  plt.plot(x,y,'ro')
  plt.plot(x,z,'black')
  plt.show()

  error=la.norm(y-z)
  print("El error es ",error)
  return Params
```

```python
Pa=Ec_normales(40)
```

```python
def QuitarDispersos(x,y):
  n=len(y)
  for i in range(n):
    for j in range(n):
      a1=np.array([x[i],y[i]])
      a2=np.array([x[j],y[j]])
      distancia=la.norm(a1-a2)
      if distancia!=0.0:
        print(distancia)
```

```python
x,y=gen_data(20)
QuitarDispersos(x,y)
```
